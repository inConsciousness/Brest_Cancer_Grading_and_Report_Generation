```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('E:/ANN Project/data/breast-cancer.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(['id','Unnamed: 32'],axis=1)
df.head()
```

```python
df["diagnosis"]=df["diagnosis"].map({'B':0,'M':1}).astype(int)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Basic info and statistics
print("Dataset Info:")
df.info()

print("\nStatistical Summary:")
print(df.describe())

# Diagnosis count plot
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='diagnosis')
plt.title('Count of Diagnosis (0 = Benign, 1 = Malignant)')
plt.show()

# Correlation heatmap
plt.figure(figsize=(15, 12))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Pairplot of a few important features
selected_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'diagnosis']
sns.pairplot(df[selected_features], hue='diagnosis')
plt.suptitle('Pairplot of Key Features by Diagnosis', y=1.02)
plt.show()

# Boxplot for feature distributions
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']])
plt.title('Feature Distributions')
plt.show()
```

```python
X = df.iloc[: , 1:]
y = df.iloc[: , 0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import ReLU,LeakyReLU,PReLU,ELU
```

```python
ann = Sequential()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
X.shape
```

```python
ann.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=X.shape[1]))
ann.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))

```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.summary()
```

```python
reg = ann.fit(X_train,y_train,validation_split=0.1,batch_size=10,epochs=50)
```

```python
reg.history
```

```python
import matplotlib.pyplot as plt

# Accuracy plot
plt.plot(reg.history['accuracy'], label='Train Accuracy')
plt.plot(reg.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Model Accuracy')
plt.legend()
plt.show()

# Loss plot
plt.plot(reg.history['loss'], label='Train Loss')
plt.plot(reg.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Model Loss')
plt.legend()
plt.show()
```

```python
y_pred = ann.predict(X_test)
y_pred = y_pred > 0.5
```

```python
test_loss, test_accuracy = ann.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")
```

```python
import pandas as pd
import io
import ipywidgets as widgets
from IPython.display import display

# File upload widget
uploader = widgets.FileUpload(accept='.csv', multiple=False)
display(uploader)

# Function to load uploaded CSV
def handle_upload(change):
    for filename, file_info in uploader.value.items():
        content = io.StringIO(file_info['content'].decode('utf-8'))
        uploaded_df = pd.read_csv(content)
        print("Uploaded File Preview:")
        display(uploaded_df.head())
        global user_input_df
        user_input_df = uploaded_df  # save globally for model prediction
```

```python
uploader.observe(handle_upload, names='value')
```

```python
from keras.models import load_model
from sklearn.preprocessing import StandardScaler
import numpy as np

# Load saved model
model = load_model('ann_model/saved_model/model_ann.h5')

# Preprocess user input
scaler = StandardScaler()
X_input = scaler.fit_transform(user_input_df)  # assumes same structure as training

# Predict
pred_probs = model.predict(X_input)
pred_labels = (pred_probs > 0.5).astype(int)

print("Predictions (0 = Benign, 1 = Malignant):")
print(pred_labels)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load Phi model
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-2")
model_phi = AutoModelForCausalLM.from_pretrained("microsoft/phi-2")

# Construct prompt for report generation
def generate_report(prob, label):
    prompt = (
        f"The model has predicted a {'malignant' if label == 1 else 'benign'} tumor "
        f"with a probability of {prob:.2f}. Generate a medical report explaining this "
        f"to a patient and recommend next steps."
    )
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model_phi.generate(**inputs, max_new_tokens=150)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Generate and display reports
for prob, label in zip(pred_probs, pred_labels):
    report = generate_report(prob[0], label[0])
    print("\n📄 AI-Generated Medical Report:")
    print(report)
```
